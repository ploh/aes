### Aufgabe 2 - Higher-Order Functions und Closures

1. Schreibe eine Funktion `make_power(exponent)`, die eine neue Funktion zurückgibt.
2. Die zurückgegebene Funktion soll eine Zahl auf den vorgegebenen Exponenten potenzieren.
3. Schreibe zusätzlich eine Funktion `apply(values, transformer)`.
4. Transformiere damit eine Liste von Zahlen.
5. Teste beides mit einer Quadrat- oder Kubikfunktion.

```python
def make_power(exponent):
    def power(value):
        return value ** exponent

    return power

def apply(values, transformer):
    return [transformer(value) for value in values]

numbers = [1, 2, 3, 4, 5]

square = make_power(2)
cube = make_power(3)

print(apply(numbers, square))
print(apply(numbers, cube))
```

# 3.7 Exceptions und Fehlerbehandlung

## Fehlerbehandlung (39)

```python
inputs = ["25", "0", "abc"]

for raw_value in inputs:
    try:
        number = int(raw_value)
        result = 100 / number
        print(f"100 / {raw_value} -> {result}")
    except ValueError as error:
        print(f"{raw_value} ist keine ganze Zahl: {error}")
    except ZeroDivisionError as error:
        print(f"Division durch 0: {error}")
```

```python
def parse_discount(raw_value):
    try:
        discount = int(raw_value)
        if discount < 0:
            raise ValueError("Rabatt darf nicht negativ sein")
    except ValueError as error:
        print(f"Fehler bei {raw_value!r}: {error}")
        raise
    finally:
        print("Prüfung beendet")

    print(f"Gültiger Rabatt: {discount}%")
    return discount


for raw_value in ["15", "-3", "x"]:
    try:
        parse_discount(raw_value)
    except Exception as error:
        message = "Ungültiger Rabatt, bitte korrigieren."
        print(message)
        # raise RuntimeError(message) from error
    print("---")
```

## Übungen zu 3.7

### Aufgabe 1 - Sichere Division

1. Schreibe eine Funktion `safe_divide(a, b)`.
2. Gib bei erfolgreicher Division das Ergebnis zurück.
3. Fange `ZeroDivisionError` ab und gib stattdessen `None` zurück.
4. Teste die Funktion mit mehreren Wertepaaren.

```python
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return None

tests = [(10, 2), (5, 0), (9, 3)]

for left, right in tests:
    print(left, right, safe_divide(left, right))
```
