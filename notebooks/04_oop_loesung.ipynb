## Attribute, Konstruktoren und `self` (5)

````python
class Article:
    tax_rate = 0.19

    def __init__(self, name, net_price):
        self.name = name
        self.net_price = net_price

    def gross_price(self):
        return self.net_price * (1 + self.tax_rate)

    def __str__(self):
        return f"{self.name}: {self.net_price} EUR (net), {self.gross_price():.2f} EUR (gross)"

    @classmethod
    def increase_tax_rate(cls, value):
        cls.tax_rate += value

    @staticmethod
    def is_valid_price(value):
        return value >= 0


coffee = Article("Coffee Beans", 12.5)
tea = Article("Green Tea", 8.0)
print(coffee)
print(tea)
````

````python
# Klassenattribut ändern
Article.increase_tax_rate(0.01)
print(coffee)
print(tea)
````

````python
# Instanzattribut erzeugen
tea.tax_rate = 0.07
print(tea)
````

````python
# Klassenattribut bleibt unverändert
print(coffee)
````

````python
# tea hat immer noch sein eigenes Instanzattribut
Article.tax_rate = 0.25
print(coffee)
print(tea)
````

````python
print(coffee.tax_rate)
print(tea.tax_rate)
print(Article.tax_rate)
````

````python
# Instanzattribut löschen
del tea.tax_rate
print(tea)
````

## Referenzen und Kopien (6)

````python
import copy

class PackingList:
    def __init__(self, items):
        self.items = items

original = PackingList(["notebook", "charger"])
shallow_copy = copy.copy(original)
deep_copy = copy.deepcopy(original)

print("original:", original.items)
print("shallow_copy:", shallow_copy.items)
print("deep_copy:", deep_copy.items)
````

````python
original.items.append("water bottle")
shallow_copy.items.append("snacks")

print("original:", original.items)
print("shallow_copy:", shallow_copy.items)
````

````python
deep_copy.items.append("jacket")

print("original:", original.items)
print("deep_copy:", deep_copy.items)
````

````python
print("shallow_copy is original:", shallow_copy is original)
print("shallow shares list:", shallow_copy.items is original.items)
print("deep shares list:", deep_copy.items is original.items)
````

### Aufgabe 1 - Klassenattribute und Objektzustand

1. Schreibe eine Klasse `Course`.
2. Jedes Objekt hat `title` und eine Liste `participants`.
3. Alle Kurse teilen das Klassenattribut `semester = "WS 2026/27"`.
4. Implementiere `enroll(name)` und `participant_count()`, sowie `__str__()`.
5. Erzeuge zwei Kurse und zeige, dass Teilnehmerlisten getrennt, `semester` aber gemeinsam ist.

````python
class Course:
    semester = "WS 2026/27"

    def __init__(self, title):
        self.title = title
        self.participants = []

    def enroll(self, name):
        self.participants.append(name)

    def participant_count(self):
        return len(self.participants)

    def __str__(self):
        return f"{self.title} ({self.semester}): {self.participant_count()} participants"

python_course = Course("Python Programming")
python_course.enroll("Alice")
python_course.enroll("Bob")
print(python_course)

law_course = Course("Introduction to Law")
law_course.enroll("Charlie")
print(law_course)
````

## Dynamisches Binden (9)

````python
class NotificationChannel:
    pass

class EmailChannel(NotificationChannel):
    def deliver(self, message: str) -> str:
        return f"email:{message.lower()}"

class SmsChannel(NotificationChannel):
    def deliver(self, message: str) -> str:
        return f"sms:{message[:10]}"

channels = [EmailChannel(), SmsChannel()]
for channel in channels:
    print(channel.deliver("System maintenance at 18:00"))
````

## Vererbung und super (10)

````python
class Person:
    def __init__(self, name):
        self.name = name

class Tutor(Person):
    def __init__(self, name, subject):
        super().__init__(name)
        self.subject = subject

    def introduce(self):
        return f"{self.name} teaches {self.subject}"

tutor = Tutor("Ada", "Python")
print(tutor.introduce())
print(isinstance(tutor, Person))
````

## Mehrfachvererbung und MRO (11)

````python
class A:
    def describe(self):
        return "A"

class B(A):
    def describe(self):
        return "B -> " + super().describe()

class C(A):
    def describe(self):
        return "C -> " + super().describe()

print(B().describe())
print(C().describe())
````

````python
class D(B, C):
    def describe(self):
        return "D -> " + super().describe()

print([cls.__name__ for cls in D.mro()])
````

````python
print(D().describe())
````

## Interfaces (12)

````python
from abc import ABC, abstractmethod

class Formatter(ABC):
    @abstractmethod
    def format(self, text: str) -> str:
        raise NotImplementedError

    def send(self, text: str):
        print(f"Sending: {self.format(text)}")

class UpperFormatter(Formatter):
    def format(self, text: str) -> str:
        return text.upper()

formatter = UpperFormatter()
formatter.send("Objekte und Methoden")
````

````python
try:
    Formatter()
except TypeError as error:
    print("ERROR", error)
````

## Overriding und Overloading (13)

````python
## Overriding

class Notification:
    def render(self) -> str:
        return "generic notification"

class EmailNotification(Notification):
    def render(self) -> str:
        return "email notification"

print(Notification().render())
print(EmailNotification().render())
````

````python
emailer = EmailNotification()
print(super(type(emailer), emailer).render())
````

````python
## "Overloading" geht nicht

class BrokenNameFormatter:
    def format_name(self, name):
        return name

    # "Destroys" the previous method, it is not an overload!
    def format_name(self, first_name, last_name):
        return f"{first_name} {last_name}"

broken = BrokenNameFormatter()
try:
    print(broken.format_name(name="Ada"))
except TypeError as error:
    print("ERROR", error)
````

````python
## Stattdessen: Flexible Parameter

class FlexibleNameFormatter:
    def format_name(self, first_name, last_name=None, *, uppercase=False):
        full_name = first_name if last_name is None else f"{first_name} {last_name}"
        return full_name.upper() if uppercase else full_name

formatter = FlexibleNameFormatter()
print(formatter.format_name("Ada"))
print(formatter.format_name("Ada", "Lovelace"))
print(formatter.format_name("Ada", "Lovelace", uppercase=True))
````

### Aufgabe 1 - Overloading mit flexiblen Parametern

1. Schreibe eine Funktion `build_label()`.
2. Sie soll mindestens die folgenden Aufrufmuster unterstützen:
   ```python
   build_label("Ada")
      -> "Ada"
   build_label("Ada", "Lovelace")
      -> "Ada Lovelace"
   build_label("Ada", "Lovelace", role="Tutorin")
      -> "Ada Lovelace (Tutorin)"
   build_label(first_name="Ada", last_name="Lovelace")
      -> "Ada Lovelace"
   build_label("Ada", last_name="Lovelace", city="London")
      -> "Ada Lovelace | city: London"
   build_label(title="Dr.", first_name="Ada", last_name="Lovelace", skills=["Python", "Mathematik"], active=True)
      -> "Ada Lovelace | title: Dr. | skills: ['Python', 'Mathematik'] | active: True"
   ```
3. Diese Aufrufmuster sollen hingegen **nicht** funktionieren, d.h. (implizit) einen Fehler werfen:
   ```python
   build_label("Ada", "Lovelace", "Tutorin")
   build_label("Ada", "Lovelace", "London")
   build_label(last_name="Lovelace", role="Tutorin")
   ```
3. Formatiere daraus jeweils einen lesbaren String.
4. Werte beliebige weitere Zusatzinfos aus `**kwargs` aus, z.B. `title`, `city`, `skills`, `active` oder weitere freie Zusatzfelder.
5. Probiere die obigen Aufrufmuster aus und gib die Ergebnisse aus. Teste auch, dass die negativen Muster Fehler werfen.

````python
def build_label(first_name, last_name=None, *, role=None, **details):
    result = first_name
    if last_name:
        result += f" {last_name}"
    if role:
        result += f" ({role})"
    for key, value in details.items():
        result += f" | {key}: {value}"
    return result


print(build_label("Ada"))
print(build_label("Ada", "Lovelace"))
print(build_label("Ada", "Lovelace", role="Tutorin"))
print(build_label(first_name="Ada", last_name="Lovelace"))
print(build_label("Ada", last_name="Lovelace", city="London"))
print(build_label(title="Dr.", first_name="Ada", last_name="Lovelace", skills=["Python", "Mathematik"], active=True))

invalid_calls = [
    lambda: build_label("Ada", "Lovelace", "Tutorin"),
    lambda: build_label("Ada", "Lovelace", "London"),
    lambda: build_label(last_name="Lovelace", role="Tutorin"),
]

for call in invalid_calls:
    try:
        print(call())
    except TypeError as error:
        print(f"ERROR: {error}")
````

### Aufgabe 2 - Vererbung, `super()` und abstrakte Basisklassen

1. Schreibe eine abstrakte Basisklasse `Employee` mit dem Attribut `name`.
2. Definiere eine abstrakte Methode `monthly_salary()`.
3. Implementiere `SalariedEmployee` mit festem Monatsgehalt und `HourlyEmployee` mit Stundengehalt und gearbeiteter Stundenzahl.
4. Gib für eine gemischte Liste von Objekten die Monatsgehälter aus.

````python
from abc import ABC, abstractmethod

class Employee(ABC):
    def __init__(self, name):
        self.name = name

    @abstractmethod
    def monthly_salary(self) -> float:
        raise NotImplementedError

class SalariedEmployee(Employee):
    def __init__(self, name, fixed_salary):
        super().__init__(name)
        self.fixed_salary = fixed_salary

    def monthly_salary(self) -> float:
        return self.fixed_salary

class HourlyEmployee(Employee):
    def __init__(self, name, rate, hours_worked):
        super().__init__(name)
        self.rate = rate
        self.hours_worked = hours_worked

    def monthly_salary(self) -> float:
        return self.rate * self.hours_worked

team = [
    SalariedEmployee("Ada", 2400),
    HourlyEmployee("Sam", rate=24, hours_worked=120),
]

for employee in team:
    print(employee.name, employee.monthly_salary())
````
