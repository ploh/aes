## Duck-Typing (18)

```python
def first_two(values):
    return values[:2]

print(first_two("Python"))
print(first_two([10, 20, 30]))
print(first_two((True, False, True)))
```

## Container-Objekte (20)

```python
## list

numbers = [1, 2, 3, 2]
print("Index-based access:\n", numbers[1])
print("Duplicates allowed:\n", numbers)

numbers[1] = 10
numbers.append(4)
print("Mutable:\n", numbers)

numbers.append("foo")
print("Mixing types:\n", numbers)
```

```python
## tuple

point = (4, 7, 'charlie')
print("Index-based access:\n", point[2])

try:
    point[0] = 10
except TypeError as error:
    print('ERROR', error)
```

```python
## dict

student = {"name": "Ada", "age": 21}
print("Key-based access:\n", student["name"])

student["city"] = "Hannover"
student.update({"course": "Python"})
print("Mutable:\n", student)

student.update({"name": "Alice"})
print("Duplicates merged:\n", student)

ordered_tags = {"python": 1, "basics": 1,}
ordered_tags.update({"and": 1, "more": 1, "in": 1, "weird": 1, "order": 1})
print("Iteration in insertion order:\n", ordered_tags)

print("Sorted dict keys:\n", sorted(ordered_tags))
```

```python
## set

tags = {"python", "basics"}
tags.add("python")
print("Duplicates auto-removed:\n", tags)

tags.update(["and", "more", "in", "weird", "order"])
print("Iteration in undefined order:\n", tags)

try:
    tags["python"]
except TypeError as error:
    print('ERROR', error)
```

# 3.7 Exceptions und Fehlerbehandlung

## Fehlerbehandlung (39)

```python
inputs = ["25", "0", "abc"]

for raw_value in inputs:
    try:
        number = int(raw_value)
        result = 100 / number
        print(f"{raw_value} -> {result}")
    except ValueError as error:
        print(f"{raw_value} ist keine ganze Zahl: {error}")
    except ZeroDivisionError as error:
        print(f"Division durch 0: {error}")
```

```python
def parse_discount(raw_value):
    try:
        discount = int(raw_value)
        if discount < 0:
            raise ValueError("Rabatt darf nicht negativ sein")
    except ValueError as error:
        print(f"Fehler bei {raw_value!r}: {error}")
        raise
    finally:
        print("Prüfung beendet")

    print(f"Gültiger Rabatt: {discount}%")
    return discount


for raw_value in ["15", "-3", "x"]:
    try:
        parse_discount(raw_value)
    except Exception as error:
        message = "Ungültiger Rabatt, bitte korrigieren."
        print(message)
        # raise RuntimeError(message) from error
    print("---")
```
