## Namen und Variablen

```python
value = 42
alias = value

print("value: ", value, type(value))
print("alias: ", alias, type(alias))

value = "foobar"

print()
print("value: ", value, type(value))
print("alias: ", alias, type(alias))
```

## Built-in-Typen

```python
values = [42, 3.5, True, "Hello", None]

for value in values:
    print(f"{value!r:>8} -> {type(value).__name__}")
```

## Zusammengesetzte Typen

```python
numbers_list = [1, 2, 2, 3]
coordinates = (10, 20)
steps = range(5)
settings = {"theme": "light", "font_size": 12}
tags = {"python", "basics", "python"}

print(numbers_list, type(numbers_list).__name__)
print(coordinates, type(coordinates).__name__)
print(list(steps), type(steps).__name__)
print(settings, type(settings).__name__)
print(tags, type(tags).__name__)
```

```python
numbers_list.append(4)
print("List after append:", numbers_list)

print("Dict access by key:", settings["theme"])
print("Set contains no duplicates:", tags)
```

## Typ-Annotationen

```python
counter: int = 3

def greet(name: str) -> str:
    return f"Hello, {name}!"

print(greet.__annotations__)
print(counter)
print(greet("Ada"))
print(greet(123))
```

## Operatoren und einfache Expressions

```python
left_operand = 7
right_operand = 3

print("left + right =", left_operand + right_operand)
print("left - right =", left_operand - right_operand)
print("left * right =", left_operand * right_operand)
print("left / right =", left_operand / right_operand)
print("left // right =", left_operand // right_operand)
print("left % right =", left_operand % right_operand)
print("left ** right =", left_operand ** right_operand)
```

```python
counter = 1
counter += 4
expression = (counter + 2) * 3

print("counter =", counter)
print("expression =", expression)
```

## Vergleiche und Boolean-Logik

```python
small_number = 2
middle_number = 5
large_number = 9

print(small_number < middle_number < large_number)
print((small_number < middle_number) and (middle_number < large_number))
print(small_number == middle_number)
print(small_number != middle_number)
```

```python
temperature = 21
is_dry = True

is_picnic_weather = temperature >= 20 and is_dry and not temperature > 30
print(is_picnic_weather)
```

## Typ-Konversion

```python
print(int("42"))
print(float("3.5"))
print(str(99))
print(list("abc"))
print(1 + 2.5, type(1 + 2.5).__name__)
print(int(3.9))
```

```python
for text in ["123", "four"]:
    try:
        print(text, "->", int(text))
    except ValueError as error:
        print(text, "->", type(error).__name__, error)
```

## Duck-Typing

```python
def first_two(values):
    return values[:2]

print(first_two("Python"))
print(first_two([10, 20, 30]))
print(first_two((True, False, True)))
```

```python
class Countdown:
    def __init__(self, start):
        self.values = list(range(start, 0, -1))

    def __getitem__(self, index):
        return self.values[index]

print(first_two(Countdown(5)))

try:
    print(first_two(10))
except TypeError as error:
    print(type(error).__name__, error)
```
