## Duck-Typing
(Folie 18)

```python
def first_two(values):
    return values[:2]

print(first_two("Python"))
print(first_two([10, 20, 30]))
print(first_two((True, False, True)))
```

```python
class Countdown:
    def __init__(self, start):
        self.values = list(range(start, 0, -1))

    def __getitem__(self, index):
        return self.values[index]

print(first_two(Countdown(5)))
```

```python
numbers = list(range(10))

squared_numbers = [number * number for number in numbers]
even_numbers = [number for number in numbers if number % 2 == 0]

print(numbers)
print(squared_numbers)
print(even_numbers)
```

## 3.4 Kontrollfluss und Blockstruktur

## Blocks und Einrückung

```python
age = 21

if age >= 18:
    message = "adult"
    if age >= 21:
        message += " and allowed to order coffee in the US example"
else:
    message = "minor"

print(message)
```

```python
broken_example = "if True:\nprint('missing indentation')"

try:
    exec(broken_example)
except IndentationError as error:
    print(type(error).__name__, error)
```

## Verzweigungen

```python
score = 78

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
else:
    grade = "D or below"

print(grade)
```

## 3.5 Funktionen und Scope

### Funktionen

```python
def area_of_rectangle(width, height):
    return width * height

double = lambda value: value * 2

print(area_of_rectangle(4, 5))
print(double(7))
```

## Rückgabewerte und Unpacking

```python
def min_and_max(values):
    return min(values), max(values)

lowest, highest = min_and_max([4, 8, 1, 9, 3])

print(lowest)
print(highest)
```

## Argument Passing

```python
def increase(number):
    number += 1
    return number

count = 10
new_count = increase(count)

print(count)
print(new_count)
```

## Scope

```python
module_level_name = "visible globally"

def demonstrate_scope():
    local_name = "visible only inside the function"
    print(module_level_name)
    print(local_name)

demonstrate_scope()

try:
    print(local_name)
except NameError as error:
    print(type(error).__name__, error)
```

## Programme strukturieren

```python
def read_data():
    return [3, 5, 8]

def process_data(values):
    return [value * 2 for value in values]

def main():
    raw_values = read_data()
    processed_values = process_data(raw_values)
    print(processed_values)

main()
```
